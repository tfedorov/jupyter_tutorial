```python
normalized= normalize_general(ukrainian_raw_txt)
tokenized = tokenize_uk.tokenize_text(normalized)
prepared_ukrainian = []
for outer_list in tokenized:
    for inner_list in outer_list:
        no_stop_words = []
        for word in inner_list:
            if word not in stop_words_ukrainian:
                #print(word)
                no_stop_words.append(word)
            #no_stop_words.append(word)
        prepared_ukrainian.append(no_stop_words)
```

```python
from gensim.models import word2vec
```

```python
model = word2vec.Word2Vec(prepared_ukrainian, size=3, min_count = 5)
```

```python
target_similar_vectors = model.wv.similar_by_word(target_word, topn = 25)

labeled_similar = [[v[0],model.wv[v[0]][0],model.wv[v[0]][1],model.wv[v[0]][2]] for v in target_similar_vectors]
labeled_similar = np.array(labeled_similar)
```

```python
target_cell = [[target_word,model.wv[target_word][0],model.wv[target_word][1],model.wv[target_word][2]]]
labeled_target = np.append(labeled_similar,target_cell, axis = 0)
```
